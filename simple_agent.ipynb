```python
# Importing Agent Devlopment Kit Components
from google.adk.agents import Agent
from google.adk.models.google_llm import Gemini
from google.adk.runners import InMemoryRunner
from google.adk.tools import google_search
from google.genai import types

print("Successfully imported the ADK Components")
```

```python
# Configuring retry options

retry_config=types.HttpRetryOptions(
    attempts=5,
    exp_base=7,
    initial_delay=3,
    http_status_codes=[429, 500, 503, 504]
)
```

```python
# Defining the agent

root_agent = Agent(
    name="helpful_assistance" ,# Used to give a name to the agent
    model=Gemini(
        model="gemini-2.5-flash-lite",
        retry_options=retry_config
    ),
    description="Simple agent that can answer general questions.",
    instruction="You are a helpful assistant. Use google search for current information or if you are unsure about something",
    tools=[google_search]
)

print("Finished setitng up the root agent")
```

```python
# Runner is a central component in ADK that helps to bring agents to life
runner = InMemoryRunner(agent=root_agent)
print("Created the runner")
```

```python
response = await runner.run_debug(
    "What is the ADK from Google? And what languages does it support?"
)
```
